Step 5. Group the year by decades and sum the values.

***Pay attention to the Population column number, summing this column is a mistake***

```python
print(crime.columns)
```

```python
import pandas as pd
crime['Decade'] = (crime.index.year // 10) * 10
decade_sum = crime.groupby('Decade').agg({
    'Population': 'max',
    'Violent': 'sum',
    'Property': 'sum',
    'Murder': 'sum',
    'Forcible_Rape': 'sum',
    'Robbery': 'sum',
    'Aggravated_assault': 'sum',
    'Burglary': 'sum',
    'Larceny_Theft': 'sum',
    'Vehicle_Theft': 'sum'
})
print(decade_sum)
```

```python
decade_sum.head()
```

```python
# To learn more about .resample (https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.resample.html)
# To learn more about Offset Aliases (http://pandas.pydata.org/pandas-docs/stable/timeseries.html#offset-aliases)

# Uses resample to sum each decade

# Uses resample to get the max value only for the "Population" column

# Updating the "Population" column
```

```python
import pandas as pd

# Assuming 'crime' is your DataFrame

# Resample the data by decade, summing all columns except 'Population'
decade_crime = crime.resample('10YE').agg({
    'Population': 'max',
    'Violent': 'sum',
    'Property': 'sum',
    'Murder': 'sum',
    'Forcible_Rape': 'sum',
    'Robbery': 'sum',
    'Aggravated_assault': 'sum',
    'Burglary': 'sum',
    'Larceny_Theft': 'sum',
    'Vehicle_Theft': 'sum'
})

# Update the 'Population' column in the original DataFrame
crime['Population'] = crime.resample('10YE')['Population'].transform('max')

print(decade_crime)
print(crime)
```

Bonus Question: What is the most dangerous decade to live in the US?

```python
import pandas as pd
crime_rates = ['Violent', 'Murder', 'Forcible_Rape', 'Robbery', 'Aggravated_assault', 'Burglary', 'Larceny_Theft', 'Vehicle_Theft']
for rate in crime_rates:
    decade_crime[f'{rate}_Rate'] = (decade_crime[rate] / decade_crime['Population']) * 100000

# Find the decade with the highest crime rate
most_dangerous_decades = {}
for rate in crime_rates:
    max_year = decade_crime[f'{rate}_Rate'].idxmax().year - (decade_crime[f'{rate}_Rate'].idxmax().year % 10)
    most_dangerous_decades[rate] = f'{max_year}-{max_year+9}'

# Print the results
for rate, decade in most_dangerous_decades.items():
    print(f'Most dangerous decade for {rate.lower().replace("_", " ")}: {decade}')
```
